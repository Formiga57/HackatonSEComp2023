```python
file_paths, labels, class_names = keras.utils.dataset_utils.index_directory("./casos","inferred",(".csv"),None,True,123,False)

batch_size = 32
shuffle = True
label_mode = "int"

def lerMedicoes(caminho,stride:int=0, head : bool = True):
    return tf.strings.to_number(tf.strings.split(tf.strings.split(tf.io.read_file(caminho)),sep=","), out_type=tf.float32)[1::stride]

args = {"stride":10,"head":True}

path_ds = tf.data.Dataset.from_tensor_slices(file_paths)
readings_ds = path_ds.map(lambda x:lerMedicoes(x,**args).to_tensor(), num_parallel_calls=tf.data.AUTOTUNE)


if batch_size is not None:
    if shuffle:
        # Shuffle locally at each iteration
        dataset = readings_ds.shuffle(buffer_size=batch_size * 8, seed=123)
    dataset = dataset.batch(batch_size)
else:
    if shuffle:
        dataset = readings_ds.shuffle(buffer_size=1024, seed=123)

readings_ds = readings_ds.prefetch(tf.data.AUTOTUNE)

if label_mode:
        label_ds = keras.utils.dataset_utils.labels_to_dataset(
            labels, label_mode, len(class_names) if class_names else 0
        )
        readings_ds = tf.data.Dataset.zip((readings_ds, label_ds))

# O dataset final ser√°  

for i, x in enumerate(readings_ds):
    print(x[1])

```
